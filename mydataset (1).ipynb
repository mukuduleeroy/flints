```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('mydataset.csv')
```

```python
average_values = df.groupby('Year')['Global_Sales'].mean()

# Create a bar plot
average_values.plot(kind='bar', color='red', edgecolor='black')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Global_Sales')
plt.title('Average Values by Category')
plt.xticks(rotation=45) 
```

```python
main_attribute= 'Year'
attributes = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']

# Plotting the line graph
plt.figure(figsize=(10, 6))  # Set the figure size

for attribute in attributes:
    plt.plot(df[attribute], label=attribute)  # Plot each attribute

# Adding labels and title
plt.xlabel(main_attribute)  # X-axis label (could also be time or another variable)
plt.ylabel(attributes)
plt.title('Line Graph of Multiple Attributes')
plt.legend()  # Show legend to identify each line
plt.grid(True)  # Add grid for better readability

# Show the plot
plt.tight_layout()  # Adjust layout to make room for the labels
plt.show()
```
